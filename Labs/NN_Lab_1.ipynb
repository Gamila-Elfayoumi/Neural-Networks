# **Implementation of a Single Neuron**

# **Building of a Simple Neural Network Layer**

```python
import numpy as np

inputs = [0.5,0.3,0.2]

# Weights and biases for 3 neurons
weights1 =np.array([[0.4,0.7,0.2],
          [0.3,0.5,0.8],
          [0.6,0.1,0.9]])
biases1 = [0.1,0.2,0.3]

# Weights and biases that cause negative outputs
weights2 = np.array([[-0.4, -0.7, 0.2],
                    [0.3, -0.5, 0.8],
                    [-0.6, 0.1, -0.9]])

biases2 = [0.1,0.2,0.3]

## let define ReLU used as Activation fuction
## return 0 if the number is negative
## and return the number if positive
def relu(num):
  return np.maximum(0, num)


## Example (1)
## Output Layer
layer_Output1 = np.dot(inputs, weights1.T) + biases1
## Activation function output
Activation_Output1 = relu(layer_Output1)
print("Example 1:")
print("Layer output (before ReLU activation):", layer_Output1)
print("Layer output (after ReLU activation):", Activation_Output1)



## Example (2)
## Output Layer
layer_Output2 = np.dot(inputs, weights2.T) + biases2
## Activation function output
Activation_Output2 = relu(layer_Output2)
print("Example 2:")
print("Layer output (before ReLU activation):", layer_Output2)
print("Layer output (after ReLU activation):", Activation_Output2)

```

# Loss function :
Mean Square Error (MSE)


```python
y_true = np.array([1,0,1,1])
y_prediction = np.array([0.9,0.2,0.8,0.4])

## MSE
def mse (y_true, y_prediction):
  return np.mean((y_true - y_prediction) ** 2)

print("Mean Square Error:", mse(y_true,y_prediction))
```
