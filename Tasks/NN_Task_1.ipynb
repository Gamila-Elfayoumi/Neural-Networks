```python
import numpy as np

# Step 1: Define Inputs
inputs = np.array([0.2, 0.4, 0.6, 0.8])

# Step 2: Initialize Random Weights and Biases
np.random.seed(42)  # Ensures reproducibility
weights = np.random.randn(4, 4)  # 4 neurons, each with 4 input weights
biases = np.random.randn(4)  # 1 bias per neuron

# Print Weights and Biases
print("Random Weights:\n", weights)
print("\nRandom Biases:\n", biases)

# Step 3: Compute Layer Output (Before Activation)
layer_output = np.dot(inputs, weights.T) + biases

# Step 4: Define ReLU Activation Function
def relu(x):
    return np.maximum(0, x)  # Replaces negative values with 0

# Step 5: Apply ReLU Activation
activated_output = relu(layer_output)

# Step 6: Print Results
print("\nLayer output (before activation):\n", layer_output)
print("\nLayer output (after ReLU activation):\n", activated_output)
```
